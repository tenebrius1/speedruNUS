```python
import requests
import concurrent.futures
import pickle
import pandas as pd
```

## Data Query (DO NOT RUN AGAIN)

```python
# Parallel 
BASE_URL = "https://api.nusmods.com/v2/"

def getAllMods():
    res = requests.get(f"{BASE_URL}2023-2024/moduleInfo.json")
    return res.json()

def getModInfo(modCode):
    res = requests.get(f"{BASE_URL}2023-2024/modules/{modCode}.json")
    return res.json()

def getAllModuleInfo(modules):
    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = executor.map(getModInfo, modules)
    return list(results)

mods = getAllMods()
all_module_codes = [mod['moduleCode'] for mod in mods]
all_module_info = getAllModuleInfo(all_module_codes)
all_module_info  # all modules info stored here, order doesn't matte
df = pd.DataFrame(all_module_info)
pickle.dump(df, open("all_module_info.pkl", "wb"))
```

```python
cscu = df[df['gradingBasisDescription'] == 'Completed Satisfactory/Unsatisfactory']
cscu
```
